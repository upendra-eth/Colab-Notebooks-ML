# Task
Develop a number guessing game where the computer generates a random number and the user tries to guess it, receiving feedback on whether their guess is too high, too low, or correct. The game should allow multiple attempts and conclude with a winning or losing message.

## Generate a Random Number

### Subtask:
Generate a random integer that the user will try to guess. This will be the 'secret' number for the game.


**Reasoning**:
To generate a random integer as per the instructions, I need to import the `random` module and then use `random.randint()` to get a number within the specified range (1 to 100) and store it in `secret_number`.



````python
import random

secret_number = random.randint(1, 100)
print(f"A secret number has been generated between 1 and 100. It is: {secret_number}")
````

## Get User Input

### Subtask:
Prompt the user to guess the secret number and store their input as an integer.


**Reasoning**:
To prompt the user for their guess, convert it to an integer, and store it, I will use the `input()` function for user interaction and the `int()` function for type conversion within a code block.



````python
user_guess = int(input("Guess a number between 1 and 100: "))
print(f"You guessed: {user_guess}")
````

## Compare Guess to Secret

### Subtask:
Compare the user's guess to the secret number using conditional statements.


```markdown
**Reasoning**:
Now that we have both the `secret_number` and the `user_guess`, the next logical step is to compare them. I will use a series of `if-elif-else` statements to check if the `user_guess` is less than, greater than, or equal to the `secret_number`.

```

**Reasoning**:
Based on the instructions, I will now implement the `if-elif-else` statement to compare `user_guess` with `secret_number` and provide feedback to the user.



````python
if user_guess < secret_number:
    print("Your guess is too low!")
elif user_guess > secret_number:
    print("Your guess is too high!")
else:
    print("Congratulations! You guessed the correct number!")
````

## Allow Multiple Attempts

### Subtask:
Allow the user to make multiple guesses until they guess the correct number or run out of a set number of attempts.


```markdown
**Reasoning**:
To allow multiple attempts, I will initialize `max_attempts` and `attempts_taken` variables. Then, I will wrap the user input and comparison logic within a `while` loop. The loop will continue as long as the guess is incorrect and attempts remain. Inside the loop, `attempts_taken` will increment, feedback will be given, and remaining attempts will be displayed.
```

**Reasoning**:
To allow the user multiple attempts, I will initialize the maximum number of attempts and a counter for attempts taken. Then, I will wrap the user input and comparison logic inside a `while` loop. The loop will continue until the user guesses correctly or runs out of attempts, providing feedback and tracking attempts throughout the game.



````python
max_attempts = 10
attempts_taken = 0
user_guess = 0 # Initialize user_guess to ensure the while loop starts

print("Let's play the number guessing game!")

while user_guess != secret_number and attempts_taken < max_attempts:
    try:
        user_guess = int(input(f"Guess a number between 1 and 100 (Attempt {attempts_taken + 1}/{max_attempts}): "))
        attempts_taken += 1

        if user_guess < secret_number:
            print("Your guess is too low!")
        elif user_guess > secret_number:
            print("Your guess is too high!")
        else:
            print(f"Congratulations! You guessed the correct number ({secret_number}) in {attempts_taken} attempts!")
            break # Exit the loop if the guess is correct

        if user_guess != secret_number and attempts_taken < max_attempts:
            print(f"You have {max_attempts - attempts_taken} attempts left.")

    except ValueError:
        print("Invalid input. Please enter an integer.")

# After the loop, check if the user ran out of attempts without guessing correctly
if user_guess != secret_number:
    print(f"Sorry, you ran out of attempts! The secret number was {secret_number}.")
````
